## __Q5 all marks and score__

```python
marks=[]   
try:
    conn=mysql.connector.connect(user='k1895908',password='H5gCl41N',host='mysql2.nms.kcl.ac.uk',database='k1895908_lab',port='33306')
    
    cursor=conn.cursor()
    cursor.execute(""" SELECT score AS Mark, DAY(timestamp)  AS Day, MONTH(timestamp) AS Month 
    FROM mark 
    GROUP BY assessmentID,studentID""")
    
    col_names=cursor.column_names
    
    rows=cursor.fetchall()
    
    for row in rows:
        marks.append(row)

    
except mysql.connector.Error as e:
    print(e)  
finally:
    cursor.close()
    conn.close()
    
marks=pd.DataFrame(marks,columns=col_names)
marks.head()
```
